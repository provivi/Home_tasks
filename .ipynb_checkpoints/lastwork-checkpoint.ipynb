Порядок действий:
1) Генерация выборок объемом 1000 (задаем случайную величину по закону двойного распределения Вейбулла)
2) Построим гистограмму выборок и график теоретической плотности распределения случайной величины
3) Сгенерируем 1000 выборок объёма n и построим гистограммы распределений их выборочных средних для значений n(5, 10, 50)
4) Вичислим значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних
5) Нарисуем плотность соответствующего нормального распределения

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
%matplotlib inline
```

```python
#1Генерация выборок объемом 1000
k = 6.0
xm = 3.0
distrRv = sts.pareto(k, loc=0, scale=xm)
distrSmpl = distrRv.rvs(1000)
print 'Теоретическое значение: ', distrRv.mean()
print 'Теоретическое стандартное отклонение: ', distrRv.std()
```

```python
#2Построим гистограмму выборок и график теоретической плотности распределения случайной величины
plt.hist(distrSmpl, bins=50, normed=True, label='sample')
x = np.linspace(0,20,1000)
pdf = distrRv.pdf(x)
plt.plot(x, pdf, label='theoretical')
plt.legend()
plt.grid()
plt.axis([0, 10, 0, 1.5])
plt.title('The probability distribution function Pareto')
plt.xlabel('x')
plt.ylabel('pdf')
```

```python
#4Вичислим значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних
distrSmplMn = k * xm / (k - 1)
distrSmplVar = np.zeros( len(smplVlm) )
for i in xrange( len(smplVlm) ):
    distrSmplVar[i] = xm**2 * k / ( (k - 1)**2 * (k - 2) ) / smplVlm[i]
    
print 'Мат.ожидание: ' + str(distrSmplMn)
for i in xrange( len(smplVlm) ):
    print 'Дисперсия n = ' + str(smplVlm[i]) + ': ' + str(distrSmplVar[i])
```
