# Подгружаем необходимые модули

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

В качестве непрерывного распределения я взял  [экспоненциальное распределение.](https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5)

# Генерируем выборку из экспоненциального распределения

В моей программе параметр для экспоненциального распределения $$\lambda = 1$$

```python
rv = sts.expon() # параметр lambda для экспоненциального распределения положим 1. 
samples = rv.rvs(1000)
```

Поскольку $$\lambda = 1$$ то матожидание, дисперсия и стандартное отклонение экспоненциального распределения: $$\mu = \mathbb{E}X = 1, \; \sigma^2 = \mathbb{D}X = 1, \; \sigma = 1$$

# Построим гистограмму и теоретическую плотность распределения
# случайной величины

Для построения графика будем использовать метод subplots ( конструкция subplots более мощная, чем конструкция plots). Subplots создает картинку из набора подграфиков.

Первые два аргумента (1,1) задают количество строк и столбцов в полученной картинке.

figsize(8,6) - это размер картинки.

dpi - dots per inch (т.е. количество точек на дюйм). 

```python
x = np.linspace(np.min(samples) - 1, np.max(samples) + 1, 500)    
fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=100)
ax.hist(samples, density=True, label='Sample density histogram',  bins=40)
ax.plot(x, rv.pdf(x), label='Theoretical PDF')
ax.set_xlabel(r'$x$')
ax.set_ylabel('Fraction of samples')
ax.legend(loc='best')
plt.show()
```

# Далее будем создавать 10000 выборок размера n и построим 

# гистограммы выборочного среднего, чтобы увидеть разницу между 

# выборками разного размера. 

Для построения гистограммы мы генерируем 10000 значений, принадлежащих выборочному среднему из n элементов. Другими словами, пусть n = 5. Тогда генерируем выборку n = 5 из исходного экспоненциального распределения, находим среднее значение элементов выборки (т.е. получаем выборочное среднее) и кладем в массив temp. Такую процедуру повторяем 10000 раз, чтобы получилась более точная гистограмма выборочного среднего.

```python
n = [5, 10, 50, 100]    
fig, ax = plt.subplots(1, 4, figsize=(18, 5), dpi=100)
means = []
for size in n:
    temp = []
    index = n.index(size)
    for _ in range(10000):
        temp.append(np.mean(rv.rvs(size)))
    ax[index].hist(temp, density=True, label='$n={}$'.format(size), bins=40)
    ax[index].legend(loc='best')
    ax[index].set_xlabel('$x$')
    means.append(temp)
ax[0].set_ylabel('Fraction of samples')
plt.show()
```

# Сравним полученные гистограммы с теоретической плотностью распределения

```python
fig, ax = plt.subplots(4, 1, figsize=(6, 20), dpi=100)
for mean in means:
    index = means.index(mean)
    sigma = 1.0/np.sqrt(n[index])
    x = np.linspace(np.min(mean) - sigma, np.max(mean) + sigma, 500)
    ax[index].hist(mean, density=True, label=r'Sample histogram, $n={}$'.format(n[index]), bins=40)
    ax[index].plot(x, sts.norm(loc=1, scale=sigma).pdf(x), label=r'CLT approximation')
    ax[index].legend(loc='best', fontsize='small')
    ax[index].set_ylabel('Fraction of samples')
    ax[index].set_xlabel('$x$')
plt.show()
```

# Вывод

Из полученных графиков мы можем увидеть, что распределение выборочного среднего действительно достаточно неплохо описывается нормальным распределением. Причем, чем больше n, тем в большей степени гистограмма будет иметь ту же форму, что и плотность распределения. В справедливости последнего тезиса можно легко убедиться, сравнив гистограммы при n = 5 и n = 100.
