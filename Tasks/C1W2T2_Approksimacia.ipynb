# Aппроксимация функции

```python
import math
def f(x):
    f = math.sin(x/5)*math.exp(x/10) + 5*math.exp(-x/2)
    return(f)
```

```python
import numpy as np
from scipy import linalg
a = np.array([[1,1],[1,15]])
b = np.array([f(1),f(15)])
result = linalg.solve(a,b)
print(result)
```

```python
from matplotlib import pylab as plt
%matplotlib inline
```

```python
a = np.array([[1,1,1],[1,8,64],[1,15,225]])
b = np.array([f(1),f(8),f(15)])
result1 = linalg.solve(a,b)
print(result1)
```

```python
plt.plot([1,8,15],result1)
plt.show()
```

```python
a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
b = np.array([f(1),f(4),f(10),f(15)])
result2 = linalg.solve(a,b)
print(result2)
```

```python
plt.plot([1,4,10,15], result2)
plt.show()
```
