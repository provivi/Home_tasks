# Центральная предельная теорема

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

В данной работе рассматривается [Бета-распределение](https://ru.wikipedia.org/wiki/%D0%91%D0%B5%D1%82%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5*).

Мат.ожидание:  $EX = \frac{\alpha}{(\alpha + \beta)},$ (1)

Дисперсия:     $DX = \frac{\alpha\cdot\beta}{(\alpha + \beta)^2\cdot(\alpha + \beta + 1)},$ (2)

где $\alpha, \beta > 0 $ - фиксированные параметры, $x \in [0,1]$. 

```python
# Зададим параметры alpha, beta вручную.

alpha = 0.4
beta = 0.4
```

```python
beta_distrib = sts.beta(alpha, beta) # зададим бета-распределенную случайную величину
```

#### Нарисуем график функции распределения и плотности распределения

```python
x = np.linspace(0,1,100) # массив точек

fig, ax = plt.subplots(1, 2, constrained_layout = True)

# график функции бета-распределения
cdf = beta_distrib.cdf(x)
ax[0].plot(x, cdf)
ax[0].set_title('The disribution function')
ax[0].set(xlabel='x', ylabel='F(x)')

# график плотности вероятности бета-распределения
pdf = beta_distrib.pdf(x)
ax[1].plot(x, pdf)
ax[1].set(xlabel='x', ylabel='f(x)')
ax[1].set_title('The probability density function')
```

#### Построим гистрограмму выборки и теоретическую плотность бета-распределения

```python
beta_values = beta_distrib.rvs(1000) #сгенерируем из бета-распределения выборку объема 1000
```

```python
plt.hist(beta_values, density=True, histtype='stepfilled') 
plt.plot(x, beta_distrib.pdf(x), 'k-', lw=3, label='theoretical pdf')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```

#### Вычислим мат.ожидание и дисперсию бета-распределения

```python
EX = alpha/(alpha+beta)
DX = alpha*beta/((alpha+beta)**2*(alpha+beta+1))
print("EX = ", EX) 
print("DX = ", DX)
```

#### Сгенерируем 1000 выборок объёма n и построим гистограммы их выборочных средних. Поверх каждой гистограммы нарисуем плотность нормального распределения, соответствующую n.

#### n = 2

```python
n = 2
list_1=[]

# сгенерируем 1000 выборок объема 2
for i in range(1000):
    list_1.append(beta_distrib.rvs(n))
    
# посчитаем выборочные средние
mean_1 = []
for i in list_1:
    mean_1.append(np.mean(i))
    
# зададим нормально распределенную случайную величину с параметрами, соответствующими бета-распределению
norm_rv1 = sts.norm(EX, np.sqrt(DX/n))

# построим гистограмму массива выборочных средних и плотность нормального распределения при n = 2
plt.plot(x, norm_rv1.pdf(x), 'k-', lw=3, label='norm_pdf (n=2)')
plt.hist(mean_1 , density = True, histtype='stepfilled')
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```

#### n = 5

```python
n = 5
list_2=[] 

# сгенерируем 1000 выборок объема 5
for i in range(1000):
    list_2.append(beta_distrib.rvs(n))
    
# посчитаем выборочные средние
mean_2 = []
for i in list_2:
    mean_2.append(np.mean(i))
    
# зададим нормально распределенную случайную величину с параметрами, соответствующими бета-распределению
norm_rv2 = sts.norm(EX, np.sqrt(DX/n))

# построим гистограмму массива выборочных средних и плотность нормального распределения  при n = 5
plt.plot(x, norm_rv2.pdf(x), 'k-', lw=3, label='norm_pdf (n=5)')
plt.hist(mean_2 , density = True, histtype='stepfilled')
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```

```python
n = 100
list_3=[]

# сгенерируем 1000 выборок объема 100
for i in range(1000):
    list_3.append(beta_distrib.rvs(n))
    
# посчитаем выборочные средние
mean_3 = []
for i in list_3:
    mean_3.append(np.mean(i))
    
# зададим нормально распределенную случайную величину с параметрами, соответствующими бета-распределению
norm_rv3 = sts.norm(EX, np.sqrt(DX/n))

# построим гистограмму массива выборочных средних и плотность нормального распределения  при n=100
plt.plot(x, norm_rv3.pdf(x), 'k-', lw=3, label='norm_pdf (n=100)')
plt.hist(mean_3 , density = True, histtype='stepfilled')
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```
